```python
import pandas as pd
import numpy as np
```

```python
wb = pd.read_excel("SecondaryMotion.xlsx")
```

```python
wb
```

```python
ref_pt1 = wb.iloc[0][3:6].to_numpy(dtype = 'float')
ref_pt2 = wb.iloc[1][3:6].to_numpy(dtype = 'float')
ref_pt3 = wb.iloc[2][3:6].to_numpy(dtype = 'float')
v1 = (ref_pt3 - ref_pt1)
v2 = (ref_pt2 - ref_pt1)
ref_norm = np.cross(v1, v2)
ref_norm = ref_norm / np.sqrt(np.sum(ref_norm**2))
rot_data = []
for i in range(0, 6):
    pt1 = wb.iloc[i*3 + 5][3:6].to_numpy(dtype = 'float') + ref_pt1
    pt2 = wb.iloc[i*3 + 6][3:6].to_numpy(dtype = 'float') + ref_pt2
    pt3 = wb.iloc[i*3 + 7][3:6].to_numpy(dtype = 'float') + ref_pt3
    
    v1 = pt3 - pt1
    v2 = pt2 - pt1

    norm = np.cross(v1, v2)
    norm = norm / np.sqrt(np.sum(norm**2))
    #print(norm)

    rot_axis = np.cross(norm, ref_norm)
    rot_axis = rot_axis / np.sqrt(np.sum(rot_axis**2))
    rot_angle = np.arccos(np.dot(norm, ref_norm))
    #print(rot_axis)
    #print(rot_angle)

    s = np.sin(rot_angle)
    c = np.cos(rot_angle)
    t = 1 - c
    x = rot_axis[0]
    y = rot_axis[1]
    z = rot_axis[2]
    y_rot = 180 / np.pi * np.arctan2(y * s - x * z * t, 1 - (y*y + z*z) * t)
    z_rot = 180 / np.pi * np.arcsin(x * y * t + z * s)
    x_rot = 180 / np.pi * np.arctan2(x * s - y * z * t, 1 - (x*x + z*z) * t)
    rot_data.append(np.array([x_rot, y_rot, z_rot]))

print("Rotation angles about X | Y | Z")
for data in rot_data:
    print("{:2.6f}, {:2.6f}, {:2.6f}".format(data[0], data[1], data[2]))
```
