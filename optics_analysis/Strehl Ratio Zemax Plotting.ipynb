```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Filepath to the output data stored in a numpy array
filePath = 'TIM_StrehlData.npy'

# data is stored in a dictionary with the keys: 'SW' and 'LW', which are numpy arrays
# containing the Strehl ratio as a function of field (row) and wavelength (column)
data = np.load(filePath).item() 
SWdata = data['SW']
LWdata = data['LW']

numFields = len(SWdata)
numWavs = len(SWdata[0])
# Parameters used to generate data
SW1 = 240
SW2 = 317
LW1 = 317
LW2 = 420
F1 = -0.464
F2 = 0.464
wavs_SW = np.linspace(SW1,SW2,numWavs)
wavs_LW = np.linspace(LW1,LW2,numWavs)
fields = np.linspace(F1,F2,numFields)
```

```python
fig, (ax_SW, ax_LW) = plt.subplots(1,2,figsize=(20,7))

ax_SW.set_title('SW Strehl Ratio')
ax_LW.set_title('LW Strehl Ratio')

ax_SW.set_xlabel('Field (degrees)')
ax_SW.set_ylabel('Wavelength ($\mu$m)')

ax_LW.set_xlabel('Field (degrees)')
ax_LW.set_ylabel('Wavelength ($\mu$m)')

im_SW = ax_SW.imshow(SWdata, extent=[F1,F2,SW1,SW2], aspect='auto')
fig.colorbar(im_SW, ax=ax_SW);

im_LW = ax_LW.imshow(LWdata, extent=[F1,F2,LW1,LW2], aspect='auto')
fig.colorbar(im_LW, ax=ax_LW);

plt.show()
```
